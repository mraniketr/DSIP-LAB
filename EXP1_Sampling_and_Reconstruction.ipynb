```
import numpy as np
import matplotlib.pyplot as plt
```

CONTINIOUS SIGNAL

```
#USER INPUTS
F=10
A=1
```

```
t = np.linspace(0, 0.2, 1000)
plt.plot(t, A * np.sin(2*np.pi*F*t));
plt.xlabel('t');
plt.ylabel('x(t)');
plt.xlim([0,0.2]);
plt.show()
```

DT CONVERSION


```
#USER INPUTS

Fs= 80
dt = 1/Fs
n = np.arange(50);
```

```
x = A*np.sin(2 * np.pi * F * n * dt)
plt.xlabel('n');
plt.ylabel('x[n]');
plt.stem(n, x);
```

Reconstructed Signal

```
t=n/Fs
plt.plot(t, A * np.sin(2*np.pi*F*t));
plt.xlabel('t');
plt.ylabel('x(t)');
plt.xlim([0,0.2]);
plt.show()
```
